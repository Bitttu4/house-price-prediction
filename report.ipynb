# House Price Prediction using Machine Learning

## Introduction

House price prediction is a critical problem in the real estate domain. Buyers, sellers, and investors need reliable pricing information to make informed decisions. This project leverages **machine learning** — specifically **Linear Regression** — to estimate residential property prices based on key structural and locational features.

By training on a real-world dataset sourced from Kaggle, the model learns patterns from historical data to generate price estimates for unseen houses. This kind of predictive modeling is widely used in property valuation, investment planning, and mortgage assessment.

## Objectives

- Predict house prices using a supervised machine learning model
- Analyze and identify the most important features influencing house prices
- Implement and evaluate a **Linear Regression** model on real-world data
- Visualize data distributions and feature-price relationships
- Lay groundwork for future improvements using advanced models

## Dataset

| Property | Details |
|---|---|
| **Source** | Kaggle |
| **Total Records** | 4,600 |
| **File Format** | CSV (`data.csv`) |
| **Target Variable** | `price` (house sale price in USD) |

### Features Used

| Feature | Description |
|---|---|
| `sqft_living` | Living area in square feet |
| `bedrooms` | Number of bedrooms |
| `bathrooms` | Number of bathrooms |
| `floors` | Number of floors |
| `view` | Quality of view (0–4 rating) |

> **Note:** Features like `sqft_living` showed the strongest correlation with price during analysis.

## Methodology

The project follows a standard machine learning pipeline:

### 1. Data Loading
The dataset (`data.csv`) is loaded using **pandas**. Initial inspection is performed using `.head()`, `.info()`, and `.describe()` to understand structure, data types, and statistical summaries.

### 2. Data Cleaning
- Checked for and handled missing/null values
- Removed duplicate entries if any
- Verified data types for each column

### 3. Exploratory Data Analysis (EDA) & Visualization
- Histograms for feature distributions
- Scatter plots of features vs. price
- Correlation heatmap using **seaborn** to identify relationships

### 4. Feature Selection
Selected five features (`sqft_living`, `bedrooms`, `bathrooms`, `floors`, `view`) based on correlation analysis and domain knowledge.

### 5. Model Training
- Split data into **80% training** and **20% testing** sets using `train_test_split`
- Trained a **Linear Regression** model from `sklearn.linear_model`

### 6. Prediction
Used the trained model to generate price predictions on the test set. Predictions saved to `output.csv`.

### 7. Evaluation
Model performance evaluated using:
- **Mean Squared Error (MSE)**
- **Root Mean Squared Error (RMSE)** — easier to interpret as it's in the same unit as price
- Visual comparison of predicted vs. actual prices

## Results

The Linear Regression model was trained and evaluated on the test split of the dataset.

| Metric | Value |
|---|---|
| **Mean Squared Error (MSE)** | 989,906,638,552.23 |
| **Root Mean Squared Error (RMSE)** | ~$994,940 |

> The high MSE is expected given the large scale of house prices. The RMSE provides a more interpretable measure — predictions deviate by approximately $994K on average, which can be improved with feature engineering and more advanced models.

Predicted values follow the general trend of actual prices, demonstrating that the selected features carry meaningful predictive signal.

## Conclusion

The Linear Regression model successfully learned relationships between house features and sale prices. Key observations:

- **Living area (`sqft_living`)** is the most significant predictor of house price.
- **View quality** and **number of bathrooms** also contribute positively to price.
- The model captures the general pricing trend, though high variance in prices makes exact prediction challenging with a linear approach alone.

Overall, this project demonstrates the end-to-end ML pipeline — from raw data to predictions — and provides a solid baseline for further refinement.
