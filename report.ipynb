# House Price Prediction using Machine Learning

## Introduction
House price prediction is an important problem in real estate. The goal of this project is to predict house prices based on features such as area, bedrooms, bathrooms, floors, and view.

Machine learning techniques are used to estimate property prices based on historical data.

## Objective
- To predict house prices using machine learning  
- To analyze important features affecting price  
- To implement Linear Regression model  

## Dataset
- Source: Kaggle  
- Total entries: 4600  
- Features used:
  - sqft_living (area)
  - bedrooms
  - bathrooms
  - floors
  - view  
- Target variable:
  - price  

## Methodology
1. Data Loading  
2. Data Cleaning  
3. Data Visualization  
4. Feature Selection  
5. Model Training  
6. Prediction  
7. Evaluation  

## Result
The model successfully predicted house prices.

Mean Squared Error (MSE): 989906638552.2268

The predicted values are close to actual values.

## Conclusion
The model performs well in predicting house prices based on selected features. Accuracy can be improved by adding more features and using advanced models.
